```
def str_to_int(s):
  return int(''.join(str(ord(c)-32).rjust(2, '0') for c in s))
  '''str.join(sequence) returns a string in which the elements of sequence 
  have been joined by str separator'''
  '''.rjust(length, fillchar) returns a new string of given length 
  after substituting a given character in left side of original string'''

def int_to_str(n):
  return '' if n==0 else int_to_str(n//100)+chr((n%100)+32)

print(str_to_int('"Hello, world?"'))
print(int_to_str(24069767679120087798276683102))
```

```
def gcd(a,b):
  if abs(a) < abs(b):
    c = a
    a = b
    b = c
  if a == 0 and b == 0:
    raise Exception('Error: a and b cannot both be zero')
  elif b == 0:
    return abs(a)
  else:
    return gcd(b, a % b)

def xgcd(a,b):
  if a == 0 and b == 0: #exception
    raise Exception('Error: a and b cannot both be zero')
  elif b == 0: #base case
    d = abs(a)
    if a < 0:
      x = -1
    else:
      x = 1
    y = 0
    return d,x,y
  if abs(b) > abs(a):
    c = a
    a = b
    b = c
    q = a // b
    r = a % b
    d,x,y = xgcd(b,r)
    return [d,x-q*y, y]
  else:
    q = a // b
    r = a % b
    d,x,y = xgcd(b,r)
    return [d,y,x-q*y]
```
