```
import numpy as NP
import numpy.linalg as LA
import imageio as IO
import matplotlib.pyplot as PLT
import scipy.misc

F = scipy.misc.face(gray=True)
PLT.imshow(F, cmap='gray')

#F.shape shows that the image is stored as a numpy array of dimensions m×n

F1 = F[200:400, 500:700]
PLT.imshow(F1, cmap='gray')

#each entry is an integer representing how dark a pixel is (0 = black, 255 = white)

U,S,VT = LA.svd(F) #U: m×m matrix with orthonormal columns; VT: n×n matrix with orthonormal columns; S: list of singular values of F

#NP.diag returns a list into a diagonal matrix

r = len(S)
Ftest = U[:,:r].dot(NP.diag(S[:r])).dot(VT[:r]) #Ftest is the same as F
u = U[:,:r].shape[0] * U[:,:r].shape[1]
V = VT[:r].transpose()
v = V.shape[0] * V.shape[1]
p = u + r + v
print(p)
```

```
s = 500

Ftest = U[:,:s].dot(NP.diag(S[:s])).dot(VT[:s])
PLT.imshow(Ftest, cmap='gray')

u = U[:,:s].shape[0] * U[:,:s].shape[1]
V = VT[:s].transpose()
v = V.shape[0] * V.shape[1]
print("Length:", u + s + v)
print("Percentage: {0}%".format((u + s + v)/p*100))
```
