```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib inline

x = np.arange(-1,1,0.1)
y = np.arange(-1,1,0.1)
#Plot the vector field for problem 1.

X, Y = np.meshgrid(x, y)
i = Y**2 - 2*X*Y
j = 3*X*Y - 6*X**2

fig, ax = plt.subplots(figsize=(10, 10))
ax.quiver(X, Y, i, j)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')

plt.show()
```

```python
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)


#Plot the first vector field for problem 3, and also plot the contour.
#To make the contour smoother, just before you plot it (but after plotting the vector field),
#do
#x = np.arange(-2,2.2,0.01)
#y = np.arange(-2,2.2,0.01)

X, Y = np.meshgrid(x, y)
z = np.log(1 + X**2 + 2*Y**2)
dx, dy = np.gradient(z)

fig, ax = plt.subplots(figsize=(10, 10))
ax.quiver(X, Y, dx, dy)
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')

x = np.arange(-2,2.2,0.01)
y = np.arange(-2,2.2,0.01)
X, Y = np.meshgrid(x, y)
z = np.log(1 + X**2 + 2*Y**2)
dx, dy = np.gradient(z)
ax.contour(X, Y, z)

plt.show()
```

```python
#For this last problem, see
#https://matplotlib.org/3.1.1/gallery/mplot3d/quiver3d.html
#The code there is copied here for your convenience.
#In this cell and the next three cells, plot the vector fields you are asked to plot.
#You only need to change the lines defining u, v, and w.


from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

# Make the direction data for the arrows
u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

plt.show()
```
