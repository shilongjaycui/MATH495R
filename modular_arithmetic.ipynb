Problem 1

```
def num_points(a,b,p):
  s = 0
  for x in range(p): #incorrect format: for x < p
    for y in range(p): #incorrect format: for y < p
      if (y**2 - x**3 - a*x - b) % p == 0:
        s += 1
  return s

print(num_points(1,1,5))
print(num_points(0,1,97))
```

```
def num_points2(a,b,c,d,e,f,p):
  s = 0
  for x in range(p): #incorrect format: for x < p
    for y in range(p): #incorrect format: for y < p
      if y**2 % p == (a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x + f) % p: #don't forget to do % p
        s += 1
  return s

def no_solution(): #a function can have no inputs
  for a in range(1,11):
    for b in range(1,11):
      for c in range(1,11):
        for d in range(1,11):
          for e in range(1,11):
            for f in range(1,11):
              if num_points2(a,b,c,d,e,f,11) == 0:
                return a,b,c,d,e,f
              
print(no_solution())
```

Problem 3

```
import sympy as sp

for n in range(3,501): #range(x,y) includes x but not y
  pseudo = 0
  if (2**(n-1) - 1) % n == 0:
    print(n)
    if sp.isprime(n) == False:
      pseudo = n
      print('The composite number is', pseudo)
```

These n are mostly prime numbers, but there's a "pseudo-prime" (341) that acts like prime numbers but is in fact composite numbers. 

Problem 4

```
for p in range(21,100):
  if (2**(p-1) - 1) % p == 0:
    for n in range(1, p):
      k = 1
      while (n**k - 1) % p != 0:
        k += 1
      print(p,k)
```

k is a factor of p - 1, and all factors of p - 1 are listed as k.
